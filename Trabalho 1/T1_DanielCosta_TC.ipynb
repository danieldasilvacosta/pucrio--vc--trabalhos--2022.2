```python
def scale( data, x_min, x_max, newx_min, newx_max ):

    data = data.astype( np.float32 )
    data = (data - x_min) * ( (newx_max - newx_min) / (x_max - x_min) ) + newx_min
    data[ data > newx_max ] = newx_max
    data[ data < newx_min ] = newx_min

    return data
```

```python
vol_f = scale( vol, 4000, 10000, 0, 1 )
vol_f
```

```python
%%time

import statistics

values = data.ravel()
values = values[ values > 0 ]

mean_value = statistics.mean( values )
median_value = statistics.median( values )
mode_value = statistics.mode( values )

print( f'Média: {mean_value}')
print( f'Mediana: {median_value}')
print( f'Moda: {mode_value}')
```

```python
def get_y( x ):

    n, bins, patches = plt.hist( values, bins='auto' )

    greater_bin = -1
    for index, bin in enumerate( bins ):

        if bin <= x:

            greater_bin = bins[ index + 1 ]
            index_bin = index + 1
            
    print( f'greater_bin: {greater_bin}' )
    print( f'index_bin: {index_bin}' )

    y = n[ index_bin ]
    print( f'y: {y}' )
    print()

    return y
```

```python
y_mean = get_y( mean_value )
y_median = get_y( median_value )
y_mode = get_y( mode_value )
```

```python
plt.figure( figsize=(8, 8) )
plt.hist( values, bins='auto' )
plt.scatter( 
    [mean_value, median_value, mode_value], 
    [y_mean, y_median, y_mode],
    # color = ['orange', 'pink', 'red'],
    color = 'red',
    zorder = 2 )
plt.show()
```
